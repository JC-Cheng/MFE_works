### Imputing data

```python
housing_train = train_set.drop("median_house_value", axis=1) 
housing_train_labels = train_set["median_house_value"].copy()
housing_test = test_set.drop("median_house_value", axis=1) 
housing_test_labels = test_set["median_house_value"].copy()

sample_incomplete_rows = housing_train[housing_train.isnull().any(axis=1)].head()
sample_incomplete_rows
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy="median")

# remove the text feature because median can only be calculated on numerical attributes
housing_num_train = housing_train.drop("ocean_proximity", axis=1)
housing_num_test = housing_test.drop("ocean_proximity", axis=1)

imputer.fit(housing_num_train)
imputer.statistics_
```

```python
X = imputer.transform(housing_num_train)
housing_train = pd.DataFrame(X, columns=housing_num_train.columns, index = list(housing_train.index.values))
housing_train.loc[sample_incomplete_rows.index.values]
```

```python
X_test = imputer.transform(housing_num_test)
housing_test = pd.DataFrame(X_test, columns=housing_num_test.columns, index = list(housing_test.index.values))
```

### Encoding attributes

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
housing_cat_train = train_set["ocean_proximity"]
housing_cat_test = test_set["ocean_proximity"]
housing_cat_encoded_train = encoder.fit_transform(housing_cat_train)
housing_cat_encoded_test = encoder.fit_transform(housing_cat_test)

housing_cat_encoded_train
```

```python
print(encoder.classes_)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
housing_cat_1hot_train = encoder.fit_transform(housing_cat_encoded_train.reshape(-1,1))
housing_cat_1hot_test = encoder.fit_transform(housing_cat_encoded_test.reshape(-1,1))

housing_cat_1hot_train
```

```python
housing_cat_1hot_train.toarray()
```

```python
housing_prepared_train = np.hstack((housing_train.values,housing_cat_1hot_train.toarray()) )
housing_prepared_test = np.hstack((housing_test.values,housing_cat_1hot_test.toarray()) )

# The most efficient way is to incorprate pipelines or functions in order to re-use the data preparation steps
```

### Model training (Linear Regression Example)

```python
from sklearn.linear_model import LinearRegression

# Create a baseline without any training
lin_reg = LinearRegression()
lin_reg.fit(housing_prepared_train, housing_train_labels)

# Grab some data
some_data = housing_prepared_train[:5]
some_labels = housing_train_labels[:5]
print "Predictions:", list(lin_reg.predict(some_data))

print "Labels:", list(some_labels)
```

## Take log-scale on the label & Normalize features

```python
f, ax = plt.subplots(2, 1, sharey=True, figsize=(10, 8))
ax[0].hist(housing_train_labels, bins=30)
ax[1].hist(np.log(housing_train_labels), bins=30)
```

```python
# nothing but rename

m = housing_prepared_train.mean()
v = housing_prepared_train.std()

X_train, y_train = (housing_prepared_train - m)/v, np.log(housing_train_labels)
X_test, y_test = (housing_prepared_test - m)/v, np.log(housing_test_labels)

print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

## SVM - SVR

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
```

```python
params={'C': [0.01, 0.03, 0.1, 0.3, 1, 3], 'kernel': ['rbf', 'linear']}

svr_cv = GridSearchCV(SVR(), param_grid=params, scoring=make_scorer(mean_squared_error),
                       cv=3, n_jobs=-1, verbose=2)

svr_cv.fit(X_train, y_train)
y_pred = svr_cv.predict(X_test)
```

```python
print svr_cv.best_params_, svr_cv.best_score_

print 'MSE:', mean_squared_error(y_test, y_pred)
print 'MAE:', mean_absolute_error(y_test, y_pred)
```

## Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
params={'n_estimators': [30, 100, 300], 
        'max_depth': [3, 5, 7],
        'max_features': ['auto', 'log2', 'sqrt']
       }

gbr_cv = GridSearchCV(GradientBoostingRegressor(), 
                       param_grid=params, scoring=make_scorer(mean_squared_error),
                       cv=3, n_jobs=-1, verbose=2)

gbr_cv.fit(X_train, y_train)
y_pred = gbr_cv.predict(X_test)
```

```python
print gbr_cv.best_params_, gbr_cv.best_score_

print 'MSE:', mean_squared_error(y_test, y_pred)
print 'MAE:', mean_absolute_error(y_test, y_pred)
```

After rescaling features and label, we consider SVM and gradient boosting tree regressor, and perform basic grid-search on hyperparameter space using 3-fold cross-validation. The best of gradient boosting tree model out perform the best SVM model on the test set in terms of mean square error and mean absolute error.
